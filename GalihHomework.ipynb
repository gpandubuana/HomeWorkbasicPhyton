### Question 4 - Which code is appropriate to drop duplicate values in 'App' column and remove its rows in Python?  Assume dataset was loaded as a pandas data frame variable: google_play_apps.

### Question 5 - Upon removing duplicate of App column, please run the following code to remove a row that contain 'Free' value at the 'Installs' column. You can check first with this code: 

```python
google_play_apps = google_play_apps[google_play_apps['Installs'] != 'Free']
google_play_apps['Installs'] = google_play_apps['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
google_play_apps['Installs'] =  google_play_apps ['Installs'].apply(lambda x: x.replace( ',' , '') if ',' in str(x) else x)
```
